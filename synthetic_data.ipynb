```python
n = 5000
X = np.empty(n)
beta = 0.95
sigma = 0.2
theta = 100
J = np.empty(n)
X[0] = 1
J[0] = 1 #inizializzo J[1] con 1 perchè con J[1]=0 c'erano problemi 
         # nel ciclo for perchè rimaneva bloccato nel while
dummy = 2
m1 = 1
```

```python
def generate_data(X, J, m1, beta, sigma, theta, dummy):
    
    #pvec = np.empty(n)
    pvec = np.exp((-1/70)*np.linspace(0, 5000, 5000))
    
    itp = 0
    ii = 0
    jj = 0
    tt = 0
    
    for i in range(1, n):
        
        M_m1 = m1*(1 - beta) # media di una binomiale
        u = random()
        v = random()
        
        if (u > beta):
            # 1: specie contaminante nuova
            X[i] = dummy
            J[i] = 0
            dummy = dummy + 1
            m1 = m1 + 1
            tt = tt + 1
  
        elif (u <= beta):
            #p = 1 - (theta + (dummy - M_m1)*sigma) / (theta + n - M_m1)
            #p = 1 - (theta + dummy)/n
            #pvec[itp] = p
            p = pvec[i-tt]
            itp = itp + 1
            
            if (v < p):
                # 2: specie non contaminante nuova
                X[i] = dummy
                J[i] = 1
                dummy = dummy + 1
                m1 = m1 + 1
                ii = ii + 1
                
            else:
                jj = jj + 1
                t = randint(0, i-1)
                
                while (J[t] == 0):
                    t = randint(1, i-1)
                    
                count = np.count_nonzero(X == X[t])
                
                # 3: non contaminante, freq 1
                if (count == 1):
                    # se quel valore l'ho già incontrato solo una volta
                    X[i] = X[t]
                    m1 = m1 - 1
                    J[i] = 1
                
                else:
                    # 4: non contaminante, freq > 1
                    X[i] = X[t]
                    J[i] = 1
    
    pvec = pvec[:ii]
    print('generated ', tt, 'contaminated singletons')
    print('generated ', ii, 'non contaminated singletons')
    print('generated ', jj, 'already observed species')
    print('sum should be total observations ->', ii+tt+jj+1)
    return X, m1, dummy, pvec
                
  
  # taglio di J, ovvero
  # controlla in X tutti quelli che hanno frequenza >1 e elimina i valori di J in tutti gli
  # indici dove compare quel numero
```

```python
X, m1, dummy, pvec = generate_data(X, J, m1, beta, sigma, theta, dummy)
```
